```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cuadratic(x,a,b,c):
    return a*(x**2) + b*x + c
```

```python
x = np.arange(0,1+1e-5,1e-2)

a = 1
b = 1
c = 0

y = cuadratic(x,a,b,c)

plt.plot(x,y)
```

```python
def dydt(x,y):
    return 2*a*x + b

def eulerIntegrator(x,y,dydt,dt):
    y_next = y + dydt(x,y)*dt
    return y_next

def simulate(y0, tTime, dt):
    nStateVar = len(y0)    
    # Allocate time and state variable vectors
    vtime = np.arange(0,tTime+dt,dt)
    Y = np.zeros([len(vtime),nStateVar])

    # Initial State
    Y[0,:] = y0
    y = y0

    # Simulation loop
    for i,t in enumerate(vtime[1:]):
        y = eulerIntegrator(t, y, dydt, dt)
        Y[i+1,:] = y
    
    return Y,vtime
```

```python
a = 1
b = 1
y0 = np.array([0])
len(y0)
stateVar,time = simulate(y0,1,1e-5)
    
plt.plot(time,stateVar)
plt.plot(x,y)
```
